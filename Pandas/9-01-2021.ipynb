## Interating with data-bases

```python
## In-side the sqlite file, many data-bases will be created
## sq-lite file --> my_database4.db  (upto 2GB sqlite file)
## data-base file --> mydata.sqlite
```

```python
import sqlite3
```

```python
## If the table already exists, it will make a connection or else 
## it will create a new data-base (mydata.sqlite)
con = sqlite3.connect('mydata.sqlite') 

# In the data-base, if there is a test table exists it will be dropped
con.execute("drop table if exists test") 

query = '''
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER );
'''
con.execute(query)  ## executing the query
con.commit()
```

```python
data = [('Atlanta','Georgia',1.25,6),
        ('Tallahassee','Florida',2.6,3),
        ('Sacramento','California',1.7,5)]

query = "INSERT INTO test VALUES(?,?,?,?)"

con.executemany(query,data)
con.commit()
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall()
rows
```

```python
print(cursor.description)
```

```python
[x[0] for x in cursor.description]
```

```python
df = pd.DataFrame(rows,columns=[x[0] for x in cursor.description])
df
```
