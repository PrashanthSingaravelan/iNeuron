```python
dates = pd.date_range('1/1/2000',periods=7)   # date_range
dates   # without frequency , it will be in the form of days
```

```python
dates = pd.date_range('1/1/2000',periods=5,freq='M')   # date_range with months 
dates
```

```python
dates = pd.date_range('1/1/2000',periods=5,freq='3M')   # date_range with months (skipping 3 months)
dates
```

```python
with open('pandas data-set/ex7.csv') as f:
    lines = list(csv.reader(f))
```

```python
lines
```

```python
header,values = lines[0],lines[1:]
print(header)
print(values)
```

```python
a = zip(*values)   # combining (1,1) and (2,2) and (3,3) combinations, * is just a pointer
for i in a:
    print(i)
```

```python
a = zip(values)   # combining (1,2,3) and (1,2,3) combinations
for i in a:
    print(i)
```

```python
data_dict = {   i:j for i,j in zip(header, zip(*values))   }
data_dict
```
