```python
import numpy as np
from decimal import Decimal as dc
```

```python
x = [1,2,3,4,5]   # x follows gaussian/normal distribution
mean = np.mean(x)

std = np.std(x)
std = int(std)

var = np.sqrt(np.std(x))
var = int(var)
```

```python
print("Mean : ",mean)
print("Std  : ",std)
print("Var  : ",var)
```

```python
y = []
z = 0
for i in x:
    z = (np.power((i-mean),2) / var )
    y.append(z)
print(y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
df.head()
```

```python
sns.pairplot(df,hue='species')
```

```python
print(np.mean(df['sepal_length']))
print(np.median(df['sepal_length']))
```

```python
from scipy import stats
print(stats.mode(df['sepal_length']))
```

```python
# Mean and median are almost same, do we have any outlier???
```

```python
list1 = list(df['sepal_length'])
list1.append(100)
print(np.mean(list1))
```

```python
## Now we can clearly say 100 is an outlier. Shall we take mean/median
```

```python
list1 = list(df['sepal_length'])
list1.append(100)
print(np.median(list1))
```

```python
## There is not-at all a difference in median, 
## but there is a significant difference in the mean
```

```python
df['species'].unique()
```

```python
sns.distplot(df['sepal_length'],kde=False)   ## basically a histogram
```

```python
## pdf --> smoothed version of histogram
sns.distplot(df['sepal_length'],kde=True)  ## kde --> kernel density function
```

```python
sns.FacetGrid(df,hue='species',size=8).map(sns.distplot,"sepal_length").add_legend()
```

```python
temp = df[df['species']=='setosa']
print(np.std(temp['sepal_length']))

temp = df[df['species']=='versicolor']
print(np.std(temp['sepal_length']))

temp = df[df['species']=='virginica']
print(np.std(temp['sepal_length']))
```

```python
## The above graph follows standard normal distribution
## least sd --> blue
## more sd --> green
```

```python
## sepal_length is overlapping a lot, so the normal distributions curves are not separated very well
## So with sepal_length is not good for classification.
```

```python
sns.FacetGrid(df,hue='species',size=8).map(sns.distplot,"petal_length").add_legend()
```

```python
## petal length is not overlapping a lot, so the normal distributions curves are also separated very well
## So petal length is good for classification
```
