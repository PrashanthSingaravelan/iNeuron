```python
import pymongo 
```

```python
## Creating a connection between the data-base and python
DEFAULT_CONNECTION_URL = "mongodb://localhost:27017/"
client = pymongo.MongoClient(DEFAULT_CONNECTION_URL)          # client --> communication channel
client
```

```python
# if connect=True, then the connection between the python and localhost(mongo-db) is created
```

```python
client.list_database_names()   # lets see what the existing list of DBs -
```

```python
# let's verify whether we have our database in the list or not 
def checkExistence_DB(DB_NAME, client):
    DBlist = client.list_database_names()    ## all the mongo-db data bases are stored in DBlist
    if DB_NAME in DBlist:
        print(f"DB: '{DB_NAME}' exists")
        return True
    else:
        print(f"DB: '{DB_NAME}' not yet present OR no collection is present in the DB")
        return False
```

```python
DB_NAME = 'sample_1'
_ = checkExistence_DB(DB_NAME=DB_NAME, client=client)
```

```python
DB_NAME = 'admin'
_ = checkExistence_DB(DB_NAME=DB_NAME, client=client)
```

### Trying to get the list of collections inside a db

```python
def CheckExistence_COL(collection_name,db_name,db):
    collection_list = db.list_collection_names()
    if collection_name in collection_list:     ## if collection is available
        print(f"Collection:'{collection_name}' in Database:'{db_name}' exists")
        return True
    else:    ## if collection is not available
        print(f"Collection:'{collection_name}' in Database:'{db_name}' does not exists OR \n\ no documents are present in the collection")
        return False
```

```python
database = client['sample-1']                            ## python -- mongodb
collection_name  = "studentdetails"
db_name = 'sample-1'

_ = CheckExistence_COL(collection_name, db_name, database)
```

### STEP 5: Find method in MongoDB

```python
database = client['sample1']
collection_name = "studentdetails"
collection = database[collection_name]

find_first_record = collection.find_one()

print(f"The first record of collection: \n{collection_name} is=\
\n{find_first_record}")
```

```python
## fetching all the records
all_record = collection.find()
for i in all_record:
    print(i)
```

```python
### Can a create a new collection from python also
collection_name = "iNeuron_Faculties"
faculties = database[collection_name]
```

```python
list_of_records_user_defined_id = [
    {"_id": "1",
    "companyName": "AshokLeyland",
    "Faculty": "Ratan Tata"},
    {"_id": "2",
    "companyName": "Apple",
    "Faculty": "Steve Jobs"},
]

faculties_record = faculties.insert_many(list_of_records_user_defined_id)
```
