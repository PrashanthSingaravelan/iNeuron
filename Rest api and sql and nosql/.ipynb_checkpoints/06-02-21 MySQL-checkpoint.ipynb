# Table creation

```python
mydb = connection.connect(host="localhost", user="root",database='sample1' ,passwd="1234",use_pure=True)  ## data-base name should be mentioned clearly 
print(mydb.is_connected())   # check if the connection is established

query = "CREATE TABLE studentdetails (studentid INT(10) AUTO_INCREMENT PRIMARY KEY,"\
                                                                       "firstname VARCHAR(60)," \
                                                                        "lastname VARCHAR(60) , "\
                                                                        "registration_date DATE, "\
                                                                        "class VARCHAR(60), "\
                                                                        "section VARCHAR(60))"

cursor = mydb.cursor()
cursor.execute(query)
print("Table is created successfully")
mydb.close()
```

```python
import mysql.connector as connection
import pandas as pd
import csv
```

```python
try:
    mydb = connection.connect(host='localhost',user='root',passwd='1234',use_pure=True)
    print(mydb.is_connected())    ## Checking the connection
    query = 'Create database GlassData123'   ## Creating a new data-base
    
    cursor = mydb.cursor()   # creating a cursor to execute a query
    cursor.execute(query)   # executing the query
    print("Data-base is created Succesfully")    
    
    ## Creating a connection to the data-base which is created above
    mydb = connection.connect(host='localhost',database='GlassData123',user='root',passwd='1234',use_pure=True)
    
    ## query to store the data
    query = "CREATE TABLE IF NOT EXISTS GlassData (Index_Number INT(10), "\
                            "Ri float(10,5), Na float(10,5) ,Mg float(10,5) , Al float(10,5), "\
                            "Si float(10,5), K float(10,5), Ca float(10,5), Ba float(10,5), "\
                            "Fe float(10,5), Class INT(5))"
    cursor = mydb.cursor()
    cursor.execute(query)
    print("Table is created successfully")
    
    ## read the glass.data file
    with open('19.2.MySQL//Imp files//glass.data', "r") as f:
        next(f)    ## to avoid the column name's which are in the 1st row of the glass.data file
        glass_data = csv.reader(f,delimiter = "\n")
        for i in enumerate(glass_data):
            for j in (i[1]):
                cursor.execute('INSERT INTO GlassData values ({values})'.format(values=(j)))
        print("Values are successfully inserted")
        mydb.commit()
        cursor.close()
        mydb.close()
except Exception as e:
    #mydb.close()
    print(str(e))
```
