```python
def Split(temp_df):
    city_1 = []
    city_2 = []
    for i in df['From_To']:
            a,b = i.split('_')
            city_1.append(a)
            city_2.append(b)

    ## converting the list into pandas_series
    city_1 = pd.Series(data = city_1,name = "City-1")
    city_2 = pd.Series(data = city_2,name = "City-2")

    ## Joining the two series with the original data-frame
    temp_df = city_1.to_frame().join(city_2.to_frame()).join(temp_df)
    return temp_df
```
