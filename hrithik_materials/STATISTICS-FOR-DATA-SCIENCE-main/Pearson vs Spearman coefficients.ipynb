# PEARSON COEFFICIENT VS SPEARMAN COEFFICIENT

```python
import numpy as np
seed = 11
rand = np.random.RandomState(seed)
x_incr = pd.DataFrame({"X":rand.uniform(0,10,100)})
x_incr["Line+"] = x_incr.X*2+1
x_incr["Exp+"] = np.exp(x_incr.X)
print(x_incr.head())
```

##  HEATMAP FOR PEARSON METHOD

```python
import seaborn as sns
sns.heatmap(x_incr.corr(method="pearson"), annot = True, square= True,fmt='.1f',cmap= 'coolwarm')
```

##  HEATMAP FOR SPEARMAN METHOD

```python
sns.heatmap(x_incr.corr(method="spearman"), annot = True, square= True,fmt='.1f',cmap= 'coolwarm')
```

## COMPARING COEFFICIENT VALUES OF PEARSON AND SPEARMAN WITH SAME 2 VARIABLES

```python
plt.scatter(x_incr["X"],x_incr["Line+"],c='purple')
plt.xlabel("X")
plt.ylabel("Line+")
```

Plot is linear

```python
from numpy import cov
cov5=cov(x_incr["X"],x_incr["Line+"])
cov5            
```

covariance is 16.181 it tells x and y has positive covariance

```python
from scipy.stats import pearsonr
corr5,p5= pearsonr(x_incr["X"],x_incr["Line+"])
print('pearson coefficient : %.3f'%corr5)
print('p-value : %.3f'%p5)
```

```python
from scipy.stats import spearmanr
corr6,p6= spearmanr(x_incr["X"],x_incr["Line+"])
print('Spearman coefficient : %.3f'%corr6)
print('p-value : %.3f'%p6)
```

In comparison for linear data both pearson and spearman works well

```python
plt.scatter(x_incr["X"],x_incr["Exp+"],c='purple')
plt.xlabel("X")
plt.ylabel("Cube+")
```

The plot is monotonic...means x and y increasing but not in constant rate.

```python
from numpy import cov
cov6=cov(x_incr["X"],x_incr["Exp+"])
cov6 
```

covariance is 6.66265900e+03...it shows x and y has postive covariance

```python
from scipy.stats import pearsonr
corr7,p7= pearsonr(x_incr["X"],x_incr["Exp+"])
print('pearson coefficient : %.3f'%corr7)
print('p-value : %.3f'%p7)
```

```python
from scipy.stats import spearmanr
corr6,p6= spearmanr(x_incr["X"],x_incr["Exp+"])
print('Spearman coefficient : %.3f'%corr6)
print('p-value : %.3f'%p6)
```

For this monotonic data plot pearson coffecient is 0.657 but spearman gives 1.000

So as per definition, the results say pearson is meant only for linear data...pearson only suitable for linear data.


Spearman coefficient gives 1.000 for this monotonic data

so as per definition spearman coefficient proves that it is especially meant for monotonic data but not strictly....it also works with linear data.

## CONCLUSION

Pearson is meant only for linearly increasing or decreasing data

Spearman is meant for both monotonic and linear data
