## tsv --> tab separated values

```python
df = pd.read_table('code_pandas1/test.tsv')   ### Reading a tsv file in the tsv format it-self
df.head()    # By-default the separator is (sep="\t")
```

```python
df_tsv = pd.read_csv('code_pandas1/test.tsv',sep="\t")   ### Reading a tsv file in the csv format
df_tsv    # By-default the separator is (sep=",") but we are changing the separator as \t in-order to read the table from the csv format
```

#### Separator plays a major role

```python
## without separater
df_tsv = pd.read_table('code_pandas1/ex1.csv')
df_tsv
```

```python
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',')
df_tsv
```

```python
## 'infer' --> by default 1st row will be the column column or attribute name
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header='infer')  
df_tsv
```

```python
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None)   ## Now the 1st row was no referred as columns, it is the data-set itself
df_tsv
```

```python
## Even know the header goes, manual names can provide

## Provide 5 different column names
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu','good'])  
df_tsv
```

```python
## Providing 1 extra column-name
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu','good','ji'])  
df_tsv
```

```python
## 1st column became the row index since i gave only 4 column names  (Will not raise any error)
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu'])  
df_tsv
```

```python
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu'], index_col=None)  
df_tsv
```

```python
## Now 2 columns became the row index
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=','
                      ,names=['hello','Kumaru','machan'])  
df_tsv
```

### Important

```python
## use columns (usecols) --> can pull out a specif column in-order for better visualization
df_tsv = pd.read_table('code_pandas1/ex1.csv',sep=',',header=None
                      ,names=['hello','Kumaru','machan','loosu','good']
                      ,usecols=['Kumaru'])  
df_tsv
```
