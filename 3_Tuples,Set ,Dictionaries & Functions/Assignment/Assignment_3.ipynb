```python
class Parent:
    def __init__(self,a,b,c):
        self.first_side  = a
        self.second_side = b
        self.third_side  = c

class Child(Parent):
    def __init__(self,a,b,c):
         Parent.__init__(self,a,b,c)   ## Passing a,b,c values to the parent class
    def area(self):
        self.s    = (self.first_side + self.second_side + self.third_side)/2
        self.area = (self.s*(self.s-self.first_side)*(self.s-self.second_side)*
                (self.s-self.third_side)) ** 0.5
        return self.area
```

```python
x = int(input("Enter the 1st side : "))
y = int(input("Enter the 2nd side : "))
z = int(input("Enter the 3rd side : "))

obj1 = Child(x,y,z)
print("Area of the triangle(with three sides give) : ",obj1.area())
```

```python
def filter_long_words(list1,n):
    list2 = [i for i in list1 if len(i)>n]
    return list2
```

```python
list1 = list(input("Enter the elements : ").split())
n = int(input("Enter the length : "))
print("The words greater than {} is : {}".format(n,*filter_long_words(list1,n)))
```

```python
def len_word(list1):
    list2 = [len(i) for i in list1]
    for i,j in zip(list1,list2):
        print("Length of {} : {}".format(i,j))

list1 = list(input("Enter the elements : ").split())
len_word(list1)
```

```python
def vowel_check(char):
    if (char=="a" or char=="e" or char=="i" or char=="o" or char=="u"):
        return True
    else:
        return False
```

```python
x = input("Enter the character : ")
if(vowel_check(x)): print("{} is a vowel".format(x))
else : print("{} is a consonant".format(x))
```
